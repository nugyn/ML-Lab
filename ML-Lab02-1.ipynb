```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
bikeShareFrame.hist(figsize=(20,20))
plt.show()
```

```python
bikeShareFrame.plot(kind='box', subplots=True, layout=(4,4), sharex=False, sharey=False, figsize=(20,20))
plt.show()
```

```python
import seaborn as sns

correlations = bikeShareFrame.corr()

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Draw the heatmap with the mask and correct as aspect ratio
sns.heatmap(correlations, cmap="PuBu", center=0, vmin=-1, vmax=1, square=True)

# fix for mpl bug that cuts off top/bottom of seaborn viz
b, t = plt.ylim() # discover the values for bottom and top
b += 0.5 # Add 0.5 to the bottom
t -= 0.5 # Subtract 0.5 from the top
plt.ylim(b, t) # update the ylim(bottom, top) values
plt.show()
```
